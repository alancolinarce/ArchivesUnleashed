```python
import pandas as pd
```

```python
#df = pd.read_csv("/Users/alancolin/DocsLocales/HumanRights_pdf-information.csv")
dataset = "https://webdata.archive-it.org/ait/files/download/ARCHIVEIT-01068/PdfInformationExtraction/pdf-information.csv.gz?access=BCZAI5TX55KXNIAHEZN36TDIOISFGXIB"
df = pd.read_csv(dataset, compression='gzip')
```

```python
#Revisa que se importó bien la base de datos

df.head()
```

```python
#Total de filas

df.shape
```

```python
#Elimina filas con urls duplicadas

df.drop_duplicates(subset="url", keep=False, inplace=True)
```

```python
#Total de filas después de eliminar duplicados

df.shape
```

```python
#Convierte a minúsculas los nombres de los pdfs

df["filename"] = df["filename"].str.lower()
```

```python
#Convierte las celdas de filename vacías a "untitled_pdf"

df["filename"] = df["filename"].fillna('untitled_pdf')
```

```python
#Cuenta cuáles son los 20 nombres de archivo más frecuentes

df['filename'].value_counts()[:20]
```

```python
#Código para buscar varias palabras en varias columnas

df[df["url"].str.contains("feminicidio|femicidio") | df["filename"].str.contains("feminicidio|femicidio")]
```

```python
df[df["url"].str.contains("feminicide|femicide") | df["filename"].str.contains("feminicide|femicide")]
```

```python
paises_AL = "mexico|guatemala|honduras|salvador|costa|panama|nicaragua|colombia|peru|venezuela|ecuador|bolivia|paraguay|argentina|chile|uruguay|cuba|dominican|puerto"
temas_a_buscar_EN = "abortion|violence|femicide|feminicide|harassment|assault|feminis|latin|gender|women"

#Una especie de búsqueda booleana. Busca en la columna de url que aparezca tanto un país de AL como un
#tema de los que nos interesa analizar. Limitante: solo se puede buscar en una columna a la vez

df[df["url"].str.contains(paises_AL) & df["url"].str.contains(temas_a_buscar_EN)]
```

```python
df[df["filename"].str.contains(paises_AL) & df["filename"].str.contains(temas_a_buscar_EN)]
```
