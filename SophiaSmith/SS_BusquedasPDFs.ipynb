```python
import pandas as pd
```

```python
#df = pd.read_csv("/Users/alancolin/Documents/ArchivesUnleashed/SophiaSmithCSVs/SinDuplicados_SS_pdf-information.csv")

dataset = "https://webdata.archive-it.org/ait/files/download/ARCHIVEIT-06150/PdfInformationExtraction/pdf-information.csv.gz?access=46ZVYG5R5PGRJ27NC6IS33TSD5D7G5CV"
df = pd.read_csv(dataset, compression='gzip')
```

```python
df.head()
```

```python
df.shape
```

```python
#Elimina urls repetidas

df.drop_duplicates(subset="url", keep=False, inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
#Convierte a minúsculas todo el texto en la columna de filename

df["filename"] = df["filename"].str.lower()
```

```python
df.head()
```

## Búsquedas

```python
#Entrega las filas que tienen un texto en específico en la columna url, en este caso, México, Colombia, Argentina, etc.

df[df["url"].str.contains("mexico|colombia|argentina|chile|bolivia")]
```

```python
#Busca las palabras México, Argentina, Colombia, etc. en todo el dataframe 

df[df.apply(lambda row: row.astype(str).str.contains("mexico|argentina|colombia|chile|guatemala|ecuador|peru", case=False).any(), axis=1)]
```

```python
#Busca cierto texto en la columna de filename (por ejemplo: violence) y entrega las filas que cumplen con el criterio

df[df["filename"].str.contains("violence")==True]
```

```python
df[df.apply(lambda row: row.astype(str).str.contains("femicide|feminicide|feminicidio", case=False).any(), axis=1)]
```

```python
#En comparación hay 225 filas con la palabra abortion en el nombre del pdf y 258 filas con la misma palabra en la url

df[df["url"].str.contains("abortion")==True]
```

## Conteo de palabras

```python
##Conteo de palabras clave en la columna filename

palabras_a_buscar = ['abortion', 'violence', 'femicide', 'feminicide', 'harassment', 'assault', 'feminis', 'latin', 'color', 'patriarchy', 'rape', 'rights', 'killing', 'murder', 'trafficking', 'prostitution', 'justice', 'minorit', 'vulnerable', 'indigenous', 'gender', 'oppression', 'movement', 'female']


for palabra in palabras_a_buscar:
    count = df["filename"].str.count(palabra).sum()
    print(palabra + ": " + str(count))
```

```python
palabras_justicia = ['justice', 'crime', 'impunity', 'process', 'legal', 'law', 'rights', 'jail', 'prison', 'policy', 'protest', 'victim', 'revictimization', 'reintegration', 'children', 'girl', 'teen', 'domestic', 'minor']
for palabra in palabras_justicia:
    count = df["filename"].str.count(palabra).sum()
    print(palabra + ": " + str(count))
```

## Descarga de archivos csv

#### CSV países latinoamericanos

```python
## Exporta en csv los pdfs que contengan el texto de algún país latinoamericano en su url o filename.

paises_AL = "mexico|guatemala|honduras|salvador|costa|panama|nicaragua|colombia|peru|venezuela|ecuador|bolivia|paraguay|argentina|chile|uruguay|cuba|dominican|puerto"

pdfs_paisesAL = df[df.apply(lambda row: row.astype(str).str.contains(paises_AL, case=False).any(), axis=1)]
pdfs_paisesAL.to_csv("pdfsPaisesAL.csv", index=True)
```

#### CSV de pdfs con el texto "women" en su nombre de archivo 

```python
pdfs_women = df[df["filename"].str.contains("women")==True]
#pdfs_women.head()

pdfs_women.to_csv("pdfsWomenFilename.csv", index=True)
```

#### CSV de pdfs con el texto "gender" en todas las columnas

```python
pdfs_gender = df[df.apply(lambda row: row.astype(str).str.contains("gender", case=False).any(), axis=1)]

pdfs_gender.to_csv("pdfsGenderAll.csv", index=True)
```
